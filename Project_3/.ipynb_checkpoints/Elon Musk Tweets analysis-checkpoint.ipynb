```python
#created a new column linked to #
b1['hashtag'] = b1['Tweet Text'].apply(lambda x: re.findall(r"#(\w+)", x))
```

```python
from ekphrasis.classes.segmenter import Segmenter
```

```python
#segmenter using the word statistics from Twitter
seg_tw = Segmenter(corpus="twitter")
```

```python
#remove digits, make all lower cases
b1['Tweet Text'] = b1['Tweet Text'].astype(str).str.replace('\d+', '')
lower_text = b1['Tweet Text'].str.lower()
```

```python
b1['Tweet Text'].head()
```

```python
def remove_punctuation(words):
 new_words = []
 for word in words:
    new_word = re.sub(r'[^\w\s]', '', (word))
    if new_word != '':
       new_words.append(new_word)
 return new_words
```

```python
b1['Tweet Text']=remove_punctuation(b1['Tweet Text'])
```

```python
b1.head()
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt
```

```python
b1['tidy_tweet'] = np.vectorize(remove_pattern)(b1['Tweet Text'], "@[\w]*") 
b1.head()
```

```python
b1['tidy_tweet']
```

```python
b1['tidy_tweet'] = b1['tidy_tweet'].str.replace("[^a-zA-Z#]", " ")
b1.head(10)
```

```python
b1.tail()
```

```python
b1['tidy_tweet']
```

```python
#remove short words
b1['tidy_tweet'] = b1['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
b1.head()
```

```python
#Text Normalization
tokenized_tweet = b1['tidy_tweet'].apply(lambda x: x.split()) # tokenizing
```

```python
tokenized_tweet.head()
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
```

```python
#Now let’s stitch these tokens back together.
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
    
b1['tidy_tweet'] = tokenized_tweet
```

```python
all_words = ' '.join([text for text in tokenized_tweet])
from wordcloud import WordCloud
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
